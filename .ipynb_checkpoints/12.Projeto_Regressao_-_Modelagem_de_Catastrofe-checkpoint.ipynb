## Tratamento para Variáveis Categóricas antes do Label Encoding

<br>

### Métodos de Tratamento

#### Agrupamento de Categorias Semelhantes

> **Descrição**: Agrupar valores individuais em categorias maiores e mais gerais. Por exemplo, agrupar países em continentes, ou agrupar tipos de desastres em categorias mais amplas.

#### Redução de Cardinalidade

> **Descrição**: Manter as categorias mais frequentes e agrupar as menos frequentes em uma categoria "Outros".


```python
## Variável Origin

# Função aprimorada para mapear valores da variável Origin para categorias mais gerais
def categorize_origin(value):
    if pd.isna(value):
        return 'Unknown'
    value = value.lower()
    if any(word in value for word in ['rain', 'rainfall', 'downpour', 'shower', 'heavy rain', 'torrential rain']):
        return 'Heavy Rains'
    elif any(word in value for word in ['monsoon', 'monsoonal']):
        return 'Monsoon Rains'
    elif any(word in value for word in ['snowmelt', 'melting snow']):
        return 'Snowmelt'
    elif any(word in value for word in ['storm', 'cyclone', 'hurricane', 'typhoon', 'tropical depression']):
        return 'Storms and Cyclones'
    elif any(word in value for word in ['drought', 'heat', 'dry', 'hot', 'wave']):
        return 'Drought and Heat'
    elif any(word in value for word in ['low pressure', 'depression', 'trough']):
        return 'Low-Pressure Systems'
    elif 'el nino' in value:
        return 'El Nino'
    elif 'la nina' in value:
        return 'La Nina'
    elif any(word in value for word in ['lightning', 'fog', 'thunderstorm', 'wind']):
        return 'Other Weather Conditions'
    elif any(word in value for word in ['collision', 'explosion', 'overloading', 'accident']):
        return 'Human Activities'
    else:
        return 'Other'

# Aplicar a função para categorizar a variável Origin
dados['Origin'] = dados['Origin'].apply(categorize_origin)

# Verificar o resultado
print(dados['Origin'].value_counts())
print('\n\n====================================================================================================\n')


## Variável Associated_Types

# Função para mapear valores da variável Associated_Types para categorias mais gerais
def categorize_associated_types(value):
    if pd.isna(value):
        return 'Unknown'
    value = value.lower()
    if 'flood' in value or 'surge' in value or 'burst bank' in value:
        return 'Flood'
    elif 'slide' in value or 'avalanche' in value:
        return 'Slide'
    elif 'storm' in value or 'cyclone' in value or 'hurricane' in value or 'typhoon' in value:
        return 'Storm'
    elif 'snow' in value or 'ice' in value or 'avalanche' in value or 'severe winter conditions' in value:
        return 'Snow/Ice'
    elif 'drought' in value or 'heat wave' in value or 'water shortage' in value:
        return 'Drought/Heat Wave'
    elif 'fire' in value or 'wildfire' in value or 'explosion' in value:
        return 'Fire'
    elif 'epidemic' in value or 'intoxication' in value or 'crop failure' in value:
        return 'Epidemic'
    elif 'pollution' in value or 'oil spill' in value or 'chemical spill' in value or 'industrial accidents' in value:
        return 'Pollution'
    elif 'earthquake' in value or 'liquefaction' in value:
        return 'Earthquake'
    elif 'tsunami' in value or 'tidal wave' in value:
        return 'Tsunami'
    elif 'transport accident' in value or 'collision' in value:
        return 'Transport Accident'
    else:
        return 'Other'

# Aplicar a função para categorizar a variável Associated_Types
dados['Associated_Types'] = dados['Associated_Types'].apply(categorize_associated_types)

# Verificar o resultado
print(dados['Associated_Types'].value_counts())
print('\n\n====================================================================================================\n')


## Variável Magnitude_Scale

# Substituir valores ausentes (NaN) por "Unknown" na variável Magnitude_Scale
dados['Magnitude_Scale'].fillna('Unknown', inplace=True)

# Verificar o resultado
print('Opções para a variável Magnitude_Scale após tratamento:')
print(dados['Magnitude_Scale'].value_counts())
print('\n\n====================================================================================================\n')

dados.info()

print('\n')
```

## Aplicando Técnica de Padronização

- Aplicaremos a técnica somente nas variáveis definidas.
- As demais variáveis são do tipo categóricas transformadas em numéricas.
- Não é necessário aplicar padronização/normalização na variável alvo.

```python
# Listar as variáveis numéricas para padronização
variaveis_para_padronizar = ['Total_Deaths', 'Total_Affected']

# Criar o scaler
scaler = StandardScaler()

# Aplicar o scaler nos dados de treino e teste usando .loc de maneira apropriada
X_treino = X_treino.copy()  # Garantir que estamos trabalhando com uma cópia completa do DataFrame
X_teste = X_teste.copy()

X_treino.loc[:, variaveis_para_padronizar] = scaler.fit_transform(X_treino.loc[:, variaveis_para_padronizar])
X_teste.loc[:, variaveis_para_padronizar] = scaler.transform(X_teste.loc[:, variaveis_para_padronizar])

# Verificar o resultado da padronização
display(X_treino.head())

# Aplicar o scaler nos dados de treino reduzidos e teste reduzidos usando .loc de maneira apropriada
X_treino_reduzidos = X_treino_reduzidos.copy()  # Garantir que estamos trabalhando com uma cópia completa do DataFrame
X_teste_reduzidos = X_teste_reduzidos.copy()

X_treino_reduzidos.loc[:, variaveis_para_padronizar] = scaler.fit_transform(X_treino_reduzidos.loc[:, variaveis_para_padronizar])
X_teste_reduzidos.loc[:, variaveis_para_padronizar] = scaler.transform(X_teste_reduzidos.loc[:, variaveis_para_padronizar])

# Verificar o resultado da padronização
display(X_treino_reduzidos.head())
```

# Seleção de Variáveis Usando KBest

- Abordagem padrão e amplamente utilizada para a seleção de variáveis é a SelectKBest do sklearn.feature_selection, que permite selecionar as melhores variáveis com base em testes estatísticos.

<br>

> Aqui vamos criar o dataset com as variáveis mais importantes selecionadas as variáveis identificadas por ambos os métodos. As variáveis que aparecem em ambos os métodos são provavelmente importantes.

<br><br>

#### Dados Completos

```python
from sklearn.feature_selection import SelectKBest, f_classif
import seaborn as sns
import matplotlib.pyplot as plt

# Separando as variáveis de entrada (X) e a variável alvo (y) no conjunto de treino final
X = X_treino
y = y_treino

# Aplicando o SelectKBest para selecionar as melhores variáveis
k_best = SelectKBest(score_func=f_classif, k='all')
X_new = k_best.fit_transform(X, y)

# Visualizando as pontuações das variáveis
scores = k_best.scores_

# Criando um DataFrame com as pontuações
features = pd.DataFrame({
    'Feature': X.columns,
    'Score': scores
})

# Ordenando as variáveis pelas pontuações
features = features.sort_values(by='Score', ascending=False)

# Exibindo as pontuações
print(features)
print('\n-------------------------------------------------------------------------------------\n')

# Selecionando as 7 melhores variáveis, removendo duplicatas
selected_features = []
for feature in features['Feature']:
    if feature not in selected_features:
        selected_features.append(feature)
    if len(selected_features) == 7:
        break

print("\nMelhores variáveis:", selected_features)
print('\n-------------------------------------------------------------------------------------\n')

# Criando novos DataFrames apenas com as melhores variáveis
X_treino_selecionado = X_treino.loc[:, selected_features]
X_teste_selecionado = X_teste.loc[:, selected_features]

# Garantindo que não haja duplicatas
X_treino_selecionado = X_treino_selecionado.loc[:, ~X_treino_selecionado.columns.duplicated()]
X_teste_selecionado = X_teste_selecionado.loc[:, ~X_teste_selecionado.columns.duplicated()]

# Visualizando graficamente as pontuações das variáveis
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Feature', data=features, palette='viridis')
plt.title('Importância das Variáveis')
plt.xlabel('Score')
plt.ylabel('Variáveis')
plt.show()

# Verificando os DataFrames com as variáveis selecionadas
display(X_treino_selecionado.sample(5))
display(X_teste_selecionado.sample(5))

# Verificando as colunas
print(X_treino_selecionado.columns)
print(X_teste_selecionado.columns)
```

- Identificar e remover os outliers extremos baseados em z-score ou IQR (Interquartile Range).

```python
# Calculando o IQR
Q1 = dados_reduzidos['Total_Damage_Adjusted'].quantile(0.25)
Q3 = dados_reduzidos['Total_Damage_Adjusted'].quantile(0.75)
IQR = Q3 - Q1

# Definindo os limites para remoção de outliers extremos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrando os dados para remover outliers extremos
dados_sem_outliers = dados_reduzidos[(dados_reduzidos['Total_Damage_Adjusted'] >= lower_bound) & (dados_reduzidos['Total_Damage_Adjusted'] <= upper_bound)]

# Verificando o resultado
print(dados_reduzidos['Total_Damage_Adjusted'].describe())
print('\n\n')
print(dados_sem_outliers['Total_Damage_Adjusted'].describe())

# Mostrando a quantidade de dados antes e depois da remoção de outliers
print('\n\n')
print("Quantidade de dados antes da remoção de outliers:", len(dados_reduzidos))
print("Quantidade de dados depois da remoção de outliers:", len(dados_sem_outliers))
```

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()


## Carrega os dados

# Carregar o conjunto de dados usando H2O
data = h2o.import_file('dados/catastrofe_reduzidos.csv')


## Pré-processar os Dados:

# Dividir os dados em treino e teste
train, test = data.split_frame(ratios=[.8])

# Definir a variável de resposta e os preditores
y = "Total_Damage_Adjusted"
x = data.columns
x.remove(y)


## Treinar o Modelo com AutoML

aml = H2OAutoML(max_models=24, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
# Ver os modelos líderes
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
# Fazer previsões no conjunto de teste
pred = aml.leader.predict(test)
display(pred.head())

# Avaliar o modelo líder
mae = mean_absolute_error(test[y].as_data_frame().values, pred.as_data_frame().values)
mse = mean_squared_error(test[y].as_data_frame().values, pred.as_data_frame().values)
rmse = np.sqrt(mse)
r2 = r2_score(test[y].as_data_frame().values, pred.as_data_frame().values)
evs = explained_variance_score(test[y].as_data_frame().values, pred.as_data_frame().values)

print('Avaliação do Modelo AutoML')
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
print('Coeficiente R2:', r2)
print('Variância Explicada:', evs)

# Salvando as Métricas
modelo_automl_metrics = pd.DataFrame({
    'Nome do Modelo': ['Modelo AutoML'],
    'Tipo de Dado': ['Dados com Melhorias'],
    'Nome do Algoritmo': ['AutoML'],
    'MAE': [mae],
    'MSE': [mse],
    'RMSE': [rmse],
    'Coeficiente R2': [r2],
    'Variância Explicada': [evs]
})

print('\n==========================================================================\n')
print('ANÁLISE DE RESÍDUOS')
# Plot dos resíduos
residuos_automl = test[y].as_data_frame().values.flatten() - pred.as_data_frame().values.flatten()

# Histograma dos resíduos
plt.figure(figsize=(10, 6))
sns.histplot(residuos_automl, kde=True, color='purple')
plt.xlabel('Resíduos')
plt.title('Histograma dos Resíduos')
plt.show()

# Exibindo a tabela de métricas
display(modelo_automl_metrics)
```
