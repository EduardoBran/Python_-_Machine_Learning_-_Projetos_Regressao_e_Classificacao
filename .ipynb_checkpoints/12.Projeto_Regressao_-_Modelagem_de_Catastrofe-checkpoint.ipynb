### Contexto

- A empresa **SafeReseguard**, uma companhia de resseguros localizada no Rio de Janeiro, está buscando aprimorar sua capacidade de prever o impacto financeiro de catástrofes naturais em suas operações. Com o aumento das mudanças climáticas, a frequência e a intensidade de eventos catastróficos como furacões, inundações e terremotos têm aumentado. A SafeReseguard deseja desenvolver um modelo preditivo que possa ajudar a estimar os potenciais prejuízos financeiros decorrentes dessas catástrofes para melhor ajustar seus contratos de resseguros e garantir uma cobertura adequada.

<br>

### Problema de Negócio

> Como podemos prever o impacto financeiro de catástrofes naturais em diferentes regiões, considerando variáveis como tipo de catástrofe, localização, severidade do evento e características socioeconômicas das regiões afetadas?

## Análise Exploratória de Dados

```python
# Informações do DataFrame
print('\n\n INFO \n\n')
dados.info()
print('\n\n ------------------------------------------------------------------------------------------ \n\n')

# Verifica se há valores ausentes e duplicados
valores_ausentes = dados.isna().sum().sum() > 0
valores_duplicados = dados.duplicated().sum() > 0

# Nomes das variáveis com valores ausentes
variaveis_ausentes = dados.columns[dados.isna().any()].tolist()

# Número de linhas duplicadas
num_linhas_duplicadas = dados.duplicated().sum()

# Porcentagem de linhas duplicadas
porcentagem_linhas_duplicadas = (num_linhas_duplicadas / len(dados)) * 100

# Exibe o resultado
print("\n\nExistem valores ausentes:", valores_ausentes)
if valores_ausentes:
    print("\nVariáveis com valores ausentes:", variaveis_ausentes)
else:
    print("\nNenhuma variável possui valores ausentes.")

print("\n\nExistem valores duplicados:", valores_duplicados)
if valores_duplicados:
    print("\nNúmero de Linhas Duplicadas:", num_linhas_duplicadas)
    print("\nPorcentagem de Linhas Duplicadas: {:.2f}%".format(porcentagem_linhas_duplicadas))
else:
    print("\nNenhuma variável possui valores duplicados.")
```

### Resumo

- A análise exploratória inicial dos dados revelou que o conjunto de dados contém **8.150 entradas** e **46 variáveis**, com uma mistura de dados numéricos e categóricos.
- A análise revelou a **presença de valores ausentes** em várias colunas, mas nenhuma linha duplicada foi encontrada.
- O conjunto de dados possui uma quantidade significativa de informações, mas será necessário lidar com os valores ausentes para uma análise e modelagem mais precisa.
