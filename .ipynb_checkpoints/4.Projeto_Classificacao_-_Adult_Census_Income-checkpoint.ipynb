### Contexto

- A **"FinTech Solutions"**, uma empresa de consultoria financeira, está desenvolvendo uma nova plataforma de serviços personalizados de planejamento financeiro.
- O **objetivo** é utilizar dados demográficos e ocupacionais para oferecer recomendações financeiras precisas e personalizadas aos seus clientes. A empresa quer prever se a renda de uma pessoa excede 50K por ano, o que permitirá segmentar os clientes em diferentes grupos de renda e ajustar os serviços oferecidos a cada grupo.

- Para alcançar esse objetivo, a FinTech Solutions firmou parcerias com diversas empresas de pesquisa de mercado e institutos de estatística que coletam e fornecem dados censitários. Esses dados incluem informações sobre idade, educação, ocupação, horas trabalhadas, e rendimentos anuais, obtidos através de pesquisas e questionários realizados de forma voluntária e anônima. Com essa base de dados, a empresa pode construir modelos preditivos robustos e responder a perguntas críticas sobre os fatores que influenciam a renda.

- Por exemplo, a empresa poderá ofertar planos de investimento específicos para pessoas com alta probabilidade de rendimento anual superior a 50K, ou cursos de qualificação profissional para grupos com menor probabilidade de atingir essa faixa de renda.

<br>

### Coleta e Uso de Dados:

- Os dados coletados são armazenados em uma base de dados contendo informações sobre 10.000 pessoas diferentes. Cada registro inclui detalhes demográficos e ocupacionais, bem como a renda anual do indivíduo.

#### Como os dados são usados:

- Quando novos clientes se inscrevem na plataforma da FinTech Solutions, eles fornecem suas informações pessoais e ocupacionais através de um formulário de cadastro online. As informações coletadas incluem idade, classe de trabalho, nível de educação, estado civil, ocupação, tipo de relacionamento, raça, sexo, número de horas trabalhadas por semana, e país de origem.

#### Previsão de Renda:
- A empresa utiliza o modelo preditivo treinado para prever a probabilidade de um novo cliente atingir uma renda anual superior a 50K com base nas informações fornecidas. Essa previsão não é para informar o cliente sobre algo que ele já sabe, mas para permitir que a empresa segmente clientes e ofereça serviços personalizados.

<br>

Com base nas previsões, a empresa pode segmentar os clientes em diferentes grupos de renda. Por exemplo:

- **Clientes com alta probabilidade de renda >50K**: Oferecer planos de investimento mais agressivos, produtos financeiros premium, e consultoria personalizada.

### Explicação Detalhada da Variável fnlwgt
#### O Que é fnlwgt?
- fnlwgt significa "final weight" (peso final). Esta variável é usada para ajustar a amostra de pesquisa para que represente melhor a população total.

#### Exemplo Prático
- Imagine uma pesquisa onde queremos saber a renda média de uma cidade, mas a amostra contém mais idosos do que a população geral.

>População Geral: 20% são idosos.
<br>
>Amostra: 50% são idosos.

- Para corrigir isso, o fnlwgt dos idosos na amostra será reduzido, e o fnlwgt dos não-idosos será aumentado, de modo que a análise final represente a composição etária real da população.

### Análise Inicial

```python
# Informações do DataFrame
dados.info()

# Verifica se há valores ausentes e duplicados
valores_ausentes = dados.isna().sum().sum() > 0
valores_duplicados = dados.duplicated().sum() > 0

# Nomes das variáveis com valores ausentes
variaveis_ausentes = dados.columns[dados.isna().any()].tolist()

# Número de linhas duplicadas
num_linhas_duplicadas = dados.duplicated().sum()

# Porcentagem de linhas duplicadas
porcentagem_linhas_duplicadas = (num_linhas_duplicadas / len(dados)) * 100

# Exibe o resultado
print("\n\nExistem valores ausentes:", valores_ausentes)
if valores_ausentes:
    print("Variáveis com valores ausentes:", variaveis_ausentes)
else:
    print("Nenhuma variável possui valores ausentes.")

print("Existem valores duplicados:", valores_duplicados)
if valores_duplicados:
    print("Número de Linhas Duplicadas:", num_linhas_duplicadas)
    print("Porcentagem de Linhas Duplicadas: {:.2f}%".format(porcentagem_linhas_duplicadas))
else:
    print("Nenhuma variável possui valores duplicados.")
```

```python
# Verificar quais colunas têm valores "?"
columns_with_question_marks = dados.columns[(dados == '?').any()].tolist()

# Contar o número total de ocorrências de "?"
total_question_marks = dados.isin(['?']).sum().sum()

# Contar o número de linhas que contêm pelo menos um valor "?"
rows_with_question_marks = (dados == '?').any(axis=1).sum()

# Exibir o resultado
print("\nColunas com valores '?':", columns_with_question_marks)
print("Número total de ocorrências de '?':", total_question_marks)
print("Número total de linhas com valores '?':", rows_with_question_marks)
```

### Resumo

- Esta análise exploratória inicial fornece uma visão geral do conjunto de dados, destacando a **ausência de valores ausentes**, a **presença de valores duplicados** e a **identificação de colunas e linhas que contêm o valor "?"**, que precisa ser tratado na etapa de pré-processamento.
