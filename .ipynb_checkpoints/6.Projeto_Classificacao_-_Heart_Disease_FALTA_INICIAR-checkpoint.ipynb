# Projeto de Classificação: "Heart Disease"

<br><br>

### Contexto

- Um hospital deseja implementar um sistema para prever a presença de doenças cardíacas em pacientes com base em dados médicos. Esse sistema ajudaria a identificar pacientes em risco e permitiria intervenções precoces. O conjunto de dados contém diversas informações clínicas sobre os pacientes, como idade, sexo, nível de colesterol, pressão arterial, entre outros.

<br>

### Problema de Negócio
- Como podemos prever a presença de doenças cardíacas em pacientes com base em suas características clínicas? Quais são os fatores mais importantes que influenciam o diagnóstico de doenças cardíacas?

<br>

### Perguntas de Negócio Adicionais:

- Como diferentes tipos de dor no peito (Chest Pain Type) influenciam a probabilidade de um paciente ter doença cardíaca?

> Esta análise pode ajudar a entender quais tipos de dor no peito são mais indicativos de doenças cardíacas, auxiliando médicos a identificar sintomas críticos de forma mais precisa.

- Como a combinação de pressão arterial em repouso e nível de colesterol sérico afeta a probabilidade de doenças cardíacas?

> Investigando a interação entre esses dois fatores, podemos identificar padrões que podem ser usados para detectar pacientes em alto risco, permitindo uma melhor personalização dos tratamentos preventivos.
