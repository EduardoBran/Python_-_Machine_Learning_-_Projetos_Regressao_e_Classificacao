## Dividindo os dados em Dados de Treino e Dados de Teste
- Nós **treinamos** o modelo com **dados de treino** e **avaliamos** o modelo com **dados de teste**.

```python
# Cria um objeto separado para a variável alvo
y = dados.Total_Damage_Adjusted

# Cria um objeto separado para as variáveis de entrada
X = dados.drop('Total_Damage_Adjusted', axis=1)

# Split em dados de treino e teste sem amostragem estratificada
X_treino, X_teste, y_treino, y_teste = train_test_split(X, 
                                                        y, 
                                                        test_size=0.2, 
                                                        random_state=1234)

# Print do shape
print(X_treino.shape, X_teste.shape, y_treino.shape, y_teste.shape)
```

<br>

## Aplicando Técnica de Padronização

- Aplicaremos a técnica somente nas variáveis definidas.
- As demais variáveis são do tipo categóricas transformadas em numéricas.
- Não é necessário aplicar padronização/normalização na variável alvo.

```python
# Listar as variáveis numéricas para padronização
variaveis_para_padronizar = ['Total_Deaths', 'Total_Affected']

# Criar o scaler
scaler = StandardScaler()

# Aplicar o scaler nos dados de treino e teste usando .loc de maneira apropriada
X_treino = X_treino.copy()  # Garantir que estamos trabalhando com uma cópia completa do DataFrame
X_teste = X_teste.copy()

X_treino.loc[:, variaveis_para_padronizar] = scaler.fit_transform(X_treino.loc[:, variaveis_para_padronizar])
X_teste.loc[:, variaveis_para_padronizar] = scaler.transform(X_teste.loc[:, variaveis_para_padronizar])

# Verificar o resultado da padronização
display(X_treino.head())
```

### Criação, Treinamento, Previsão e Avaliação do Modelo
