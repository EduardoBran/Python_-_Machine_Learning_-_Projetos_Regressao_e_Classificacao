# Projeto de Regressão: "Wine Quality"

<br><br>

### Contexto

- A vinícola “Vinhos de Excelência” deseja melhorar a qualidade dos seus vinhos e otimizar o processo de produção. Eles coletaram dados sobre características físico-químicas dos vinhos, como acidez, açúcar residual, pH e teor alcoólico, bem como a qualidade dos vinhos determinada por testes sensoriais realizados por especialistas.

<br>

### Problema de Negócio
- Como podemos prever a qualidade do vinho (classificação em uma escala de 0 a 10) com base nas características físico-químicas? Quais são as características mais importantes que influenciam a qualidade do vinho?

### Perguntas de Negócio Adicionais:

- Como as interações entre diferentes características físico-químicas afetam a qualidade do vinho?

> Esta análise pode revelar combinações específicas de características, como níveis de acidez e teor alcoólico, que são mais propensas a produzir vinhos de alta qualidade. Isso ajudará a vinícola a ajustar seus processos de produção para otimizar a qualidade.

- Como variáveis externas, como a safra (ano de produção) e a região de origem, influenciam a qualidade do vinho?

> Adicionando informações sobre a safra e a região de origem ao modelo, podemos determinar se há variáveis externas que influenciam significativamente a qualidade do vinho. Isso pode ajudar na tomada de decisões sobre onde e quando produzir vinhos de alta qualidade.
