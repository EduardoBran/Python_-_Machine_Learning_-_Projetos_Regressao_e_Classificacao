### Motivo da Exclusão de Variáveis

Aqui está uma breve explicação para a exclusão de cada variável:

- *Dis No*: Número de identificação do desastre, não relevante para a análise preditiva.
- *Glide*: Identificação global do desastre, não necessária para a modelagem.
- *Disaster_Group*: Essa variável possui apenas um valor constante e assim não fornece informação diferenciada para o modelo preditivo, pois não contribui para a variabilidade dos dados.
- *Disaster_Subsubtype*: Muitos valores ausentes (apenas 496 de 4863 não são nulos), tornando-a difícil de usar de forma eficaz.
- *Event Name*: Nome do evento, considerado não relevante para a análise preditiva.
- *Country*: A variável Country oferece uma categorização mais detalhada, mas é amplamente representada pela variável Continent, que agrupa países em categorias geográficas mais amplas. Outro problema é: com 200 categorias únicas, Country pode introduzir sparsity nos dados, dificultando a modelagem e aumentando o risco de overfitting
- *ISO*: Código ISO do país, não necessário, pois a variável Continent já engloba os dados de localização.
- *Location*: A variável Location apresenta 4163 categorias únicas, o que pode introduzir muita granularidade e complexidade ao modelo, dificultando a interpretação e a generalização. Além disso, informações geográficas detalhadas já são representadas de maneira mais consolidada e útil pelas variáveis Country e Continent
- *Origin*: Muitos valores ausentes (1289 de 4863).
- *Associated_Dis* e *Associated_Dis2*: Muitos valores ausentes (1538 e 435 de 4863, respectivamente).
- *OFDA_Response*: Muitos valores ausentes (799 de 4863).
- *Appeal*: Muitos valores ausentes (1069 de 4863).
- *Declaration*: Muitos valores ausentes (1401 de 4863).
- *Latitude*: Latitude da localização do desastre, muitos valores ausentes e localização geográfica já representada por outras variáveis.
- *Longitude*: Longitude da localização do desastre, muitos valores ausentes e localização geográfica já representada por outras variáveis.
- *Local Time*: Tempo local do evento, não relevante para a análise preditiva.
- *River Basin*: Bacia hidrográfica afetada, muitos valores ausentes e considerada menos relevante.
- *Aid Contribution*: Contribuição de ajuda em milhares de dólares, muitos valores ausentes e não é diretamente relevante.
- *Dis_Mag_Value*: Muitos valores ausentes.
- *Admin1 Code*: Código administrativo 1, muitos valores ausentes e granularidade das unidades administrativas menos relevante.
- *Admin2 Code*: Código administrativo 2, muitos valores ausentes e granularidade das unidades administrativas menos relevante.
- *Adm Level*: Muitos valores ausentes (2771 de 4863), e pode ser menos relevante.
- *Geo Locations*: Localizações geográficas afetadas, muitos valores ausentes e localização já representada por outras variáveis.
- *Reconstruction Costs*: Custo de reconstrução em milhares de dólares, muitos valores ausentes.
- *No_Injured e No_Affected*: Ambas as variáveis têm uma alta proporção de valores ausentes. A presença de muitos valores extremos (outliers) torna a imputação mais desafiadora e pode não resultar em estimativas confiáveis.
- *No_Homeless*: Muitos valores ausentes, dificultando a utilização eficaz desta variável na modelagem.
- *Insured Damages*: Danos segurados em milhares de dólares, muitos valores ausentes e não diretamente relevante para o modelo preditivo focado em danos totais.
- *CPI*: Índice de Preços ao Consumidor, pode não ser diretamente relevante para a predição de danos causados por catástrofes naturais.

```python
from scipy.stats import mode

## Tratando com Imputação da Moda

# Função para imputar moda
def imputar_moda(data, columns):
    for column in columns:
        moda = mode(data[column].dropna())[0][0]
        data[column].fillna(moda, inplace=True)

# Variáveis para imputação da moda
variaveis_moda = ['Start_Month', 'Start_Day', 'End_Month', 'End_Day']
imputar_moda(dados, variaveis_moda)


## Tratando com Imputação da Mediana

# Função para imputar mediana
def imputar_mediana(data, columns):
    for column in columns:
        mediana = data[column].median()
        data[column].fillna(mediana, inplace=True)

# Variáveis para imputação da mediana
variaveis_mediana = ['Total_Deaths', 'Total_Affected']
imputar_mediana(dados, variaveis_mediana)


## Verificando Resultado

# Criar DataFrame temporário com as variáveis que possuem valores ausentes
dados_temp = dados[dados.columns[dados.isna().any()].tolist()]

# Exibir informações do DataFrame temporário
print(dados_temp.info())
print('\n\n================================================================================================\n\n')

# Exibir a quantidade de valores ausentes em cada variável
print('Total de Linhas Com Valores Ausentes Por Coluna\n\n')
print(dados_temp.isna().sum())
```

- **Duração do Desastre (Disaster_Duration)**: A duração do desastre pode ser um indicativo da gravidade e do impacto do evento. Desastres mais longos podem resultar em maiores danos totais.

#### Duração do Desastre (Disaster_Duration)

```python
def ajustar_dia(mes, dia, ano):
    # Adjust for leap years
    if mes == 2:
        if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):  # Leap year check
            if dia > 29:
                dia = 29
        elif dia > 28:
            dia = 28
    elif mes in [4, 6, 9, 11] and dia > 30:
        dia = 30
    elif dia > 31:
        dia = 31
    return pd.to_datetime(f'{ano}-{mes:02d}-{dia:02d}', errors='coerce')

# Applying the function to correct invalid dates
dados['Start_Date'] = dados.apply(lambda row: ajustar_dia(int(row['Start_Month']), int(row['Start_Day']), int(row['Year'])), axis=1)
dados['End_Date'] = dados.apply(lambda row: ajustar_dia(int(row['End_Month']), int(row['End_Day']), int(row['Year'])), axis=1)

# Ensure End_Date is not earlier than Start_Date
dados['End_Date'] = dados.apply(lambda row: max(row['End_Date'], row['Start_Date']), axis=1)

# Calculate the duration of the disaster in days
dados['Disaster_Duration_Days'] = (dados['End_Date'] - dados['Start_Date']).dt.days

# Remove intermediate columns
dados.drop(columns=['Start_Date', 'End_Date'], inplace=True)

# Display statistical summary of Disaster_Duration_Days
print(dados['Disaster_Duration_Days'].describe())

# Visualização dos outliers antes do tratamento
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
sns.boxplot(data=dados['Disaster_Duration_Days'])
plt.xticks(rotation=90)
plt.title('Boxplot para Detecção de Outliers (Antes do Tratamento)')

# Histograma antes do tratamento
plt.subplot(1, 2, 2)
sns.histplot(dados['Disaster_Duration_Days'], bins=50, kde=True)
plt.title('Histogram of Disaster Duration Days (Antes do Tratamento)')
plt.xlabel('Duration Days')
plt.ylabel('Frequency')
plt.show()

# Aplicar Winsorization
dados['Disaster_Duration_Days_Winsorized'] = winsorize(dados['Disaster_Duration_Days'], limits=[0.01, 0.01])

# Aplicar Transformação Logarítmica
dados['Disaster_Duration_Days_Log'] = np.log1p(dados['Disaster_Duration_Days_Winsorized'])

# Verificação de valores negativos ou NaN após a transformação
assert dados['Disaster_Duration_Days_Log'].isna().sum() == 0, "Existem valores NaN após a transformação logarítmica"
assert (dados['Disaster_Duration_Days_Log'] < 0).sum() == 0, "Existem valores negativos após a transformação logarítmica"

# Visualização dos outliers após Winsorization e Transformação Logarítmica
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
sns.boxplot(data=dados['Disaster_Duration_Days_Log'])
plt.xticks(rotation=90)
plt.title('Boxplot para Detecção de Outliers (Após Tratamento)')

# Histograma após Winsorization e Transformação Logarítmica
plt.subplot(1, 2, 2)
sns.histplot(dados['Disaster_Duration_Days_Log'], bins=50, kde=True)
plt.title('Histogram of Disaster Duration Days (Após Tratamento)')
plt.xlabel('Log(Duration Days)')
plt.ylabel('Frequency')
plt.show()

# Estatísticas descritivas após tratamento
dados.drop(columns=['Disaster_Duration_Days', 'Disaster_Duration_Days_Winsorized'], inplace=True)
dados['Disaster_Duration_Days'] = dados['Disaster_Duration_Days_Log']
dados.drop(columns=['Disaster_Duration_Days_Log'], inplace=True)
print(dados['Disaster_Duration_Days'].describe())
```
